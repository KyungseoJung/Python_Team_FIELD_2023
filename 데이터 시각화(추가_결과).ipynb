```python
# 학령인구 0인 데이터를 제외
# 빨간 원 : 학교 - 원의 크기 : 학령인구 수에 비례

import folium
import pandas as pd
from geopy.distance import geodesic


# 지도 생성
m = folium.Map(location=[37.437793, 126.575861], zoom_start=11, tiles="Stamen Terrain")

# 학교 위치 표시 (원)       
for idx, school in df_final_target.iterrows():
    if(school["할당 학령인구"] != 0) : 
        folium.Marker([school["Y"], school["X"]], tooltip=school["학교id"], icon=folium.Icon(color='red')).add_to(m)

# 학생 위치 표시 (원)
for idx, student in df_students.iterrows():
    folium.CircleMarker([student["lat"], student["long"]], radius= 3, color= 'blue', fill = True).add_to(m)

    
# 지도 출력
m
```
