```python
import pandas as pd

## 학생 수
df_students = pd.read_csv("./data/data2.csv", encoding="cp949")
df_schools = pd.read_csv("./searchData/초등학교_위경도.csv", encoding = "cp949")
df_schools_pre = pd.read_csv("./searchData/target_전처리_1.csv", encoding = "cp949")
```

```python
import folium
import pandas as pd
from geopy.distance import geodesic

# # 학교 위치 데이터 (예시)
# schools_data = [
#     {"name": "학교1", "lat": 37.500123, "long": 127.033456},
#     {"name": "학교2", "lat": 37.512345, "long": 127.045678}
# ]

# # 학생 위치 데이터 (예시)
# students_data = [
#     {"lat": 37.505678, "long": 127.037890},
#     {"lat": 37.518901, "long": 127.051234},
#     # ...
# ]

# 데이터프레임 생성
# schools_df = pd.DataFrame(schools_data)
# students_df = pd.DataFrame(students_data)

# 지도 생성
m = folium.Map(location=[37.437793, 126.575861], zoom_start=11,tiles="Stamen Terrain")

# 학교 위치 표시
for idx, school in df_schools.iterrows():
    folium.Marker([school["lat"], school["long"]], popup=school["name"], icon=folium.Icon(color='red')).add_to(m)

# 학생 위치 표시
for idx, student in df_students.iterrows():
    folium.CircleMarker([student["lat"], student["long"]], radius= student["초등학령인구"]/10, color= 'blue', fill = True).add_to(m)

# 학교 위치 기준으로 1.5km 반경 원 표시
for idx, school in df_schools.iterrows():
    folium.Circle(
        location=[school["lat"], school["long"]],
        radius=1500,  # 1.5km 반경을 미터로 변환
        color='red',
    ).add_to(m)

# 각 학교 위치 기준으로 1.5km 반경 내 학생 데이터 개수 확인
for idx, school in df_schools.iterrows():
    school_position = (school["lat"], school["long"])
    students_within_radius = df_students.apply(lambda student: geodesic(school_position, (student["lat"], student["long"])).kilometers <= 1.5, axis=1)
#     print(f"{school['name']} 위치 기준 1km 반경 내 학생 수: {students_within_radius.sum()}")
    total_population = df_students[students_within_radius]["초등학령인구"].sum()
    print(f"{school['name']} 위치 기준 1.5km 반경 내 학생 인구 합계: {total_population}")
    
# 지도 출력
m
```
